```python
import pandas as pd
from scipy import interpolate
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./スクショ.csv")
```

```python
dff = df[::25].reset_index(drop=True)
```

```python
# スプライン補完
spline_data = interpolate.Akima1DInterpolator(dff["x"], dff["y"])
```

```python
X = [i for i in range(249)]
Y = spline_data(x=X)
```

```python
sns.set()
sns.scatterplot(data=dff, x="x", y="y", color="red", label="original data(mabiki)")
sns.lineplot(data=df, x="x", y="y", label="original data")
sns.lineplot(x=X, y=Y, label="spline data")
plt.ylim(0, 100)
```

```python
output_data = pd.DataFrame({"X": X, "Y": Y})
```

```python
output_data.to_csv("スプライン補完データ.csv", index=False)
```
